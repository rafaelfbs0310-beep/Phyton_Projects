```python
#Exercício 4:

print("__________CALCULADORA PHYTON___________")
print("+ Adição")
print("- Subtração")
print("* Multiplicação")
print("/ Divisão")
print("Presione qualquer outra tecla pra sair.")

op= input("qual operação deseja realizar? ")
x= int(input("digite o 1º número: "))
y= int(input("digite o 2º número:"))

if(op == "+"):
  res= x + y
  print(f"resultado: {x} + {y} = {res}")

elif(op == "-"):
  res= x-y
  print(f'resultado: {x} - {y} = {res}')

elif(op == "*"):
  res= x*y
  print(f'resultado: {x} * {y} = {res}')

elif(op == "/"):
  res= x/y
  print(f'resultado: {x} / {y} = {res}')

else:
  print("Saindo...")
```
