```python
#Exercício 5:

KWh= float(input("Quantos Kwh? "))
tipo = input("Qual o tipo da instalação? (R, C ou I?) ")

if(tipo == "R"):
  if KWh >=500:
      preco = 0.65
  else:
    preco = 0.4
    print(f'Total a pagar: {KWh * preco}')

elif(tipo == "C"):
  if KWh >=1000:
      preco = 0.6
  else:
     preco = 0.55
     print(f"total a pagar: {KWh * preco}")


elif(tipo == "I"):
  if KWh >=5000:
    preco = 0.6
  else:
    preco = 0.4
    print(f"total a pagar: {KWh * preco}")

else:
  print("Instalação inválida. Encerrando....")
```
